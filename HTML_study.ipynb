# 프론트 엔드의 이해


# 1. HTML (Hyper Text Markup Language)

- 인터넷 서비스의 하나인 월드 와이드 웹을 통해 볼 수 있는 문서를 만들 때 사용하는 웹 언어의 한 종류

- 웹 사이트를 쉽게 제작하고 소스를 넣을 수 있다

- 인터넷 블로그 or 카페를 쉽게 수정할 수 있다
<br>

 **HTML 문서 구조**

 ```
 <!DOCTYPE html>
 <html>
    <head>
        <title>제가 제작한 홈페이지 입니다.</title>
    </head>

    <body>
        <p> 안녕하세요, Seongmin의 홈페이지에 오신 것을 환영합니다.
        </p>
    </body>
 </html>
  ```
<br><br>

## WWW (World Wide Web)
: 인터넷, 컴퓨터, 웹 브라우저를 통해 글자, 이미지, 동영상, 음성 등의 데이터를 사용자에게 제공하거나, 컴퓨터와 사용자 간의 의사소통을 도와주는 서비스
<br><br>

# **HTML 문서 구조** (구체적 정리)

```
<!DOCTYPE html>
• HTML 문서의 버전을 알려주는 역할
• HTML5 문서임

<html></html>
• 문서의 시작과 끝을 의미

<body></body>
• 브라우저 실행 화면에 보이는 내용의 시작과 끝을 의미

<head></head>
• HTML 문서에서 필요한 사항이나 특수한 기능을 설정할 때 사용
• <title>, <meta>, <link>, <style>, <script>태그로 HTML문서의 제목 설정,
문자 세트 설정, 외부파일연결, CSS 정의, 자바스크립트 정의 등의 역할
수행

<title></title>
• 실행 화면 왼쪽 상단의 탭 버튼에 보이는 웹 페이지 제목을 표시
```

<br>

### 1. **HTML 태그는 이름을 꺾쇠 괄호(<>)로 감싸서 표현한다.**

  ```
  <태그 이름>
  </태그 이름>
  ```

- 시작 태그와 종료 태그 한쌍으로 구성

<br>

### 2. **태그는 소문자로 작성한다.**

- HTML 태그는 대/소문자를 구별하지는 않지만
 **HTML5 표준 명세서**에서는 태그와 태그 안에 사용하는 속성들을 소문자로 작성할 것을
권장한다.

  ```
  <a href=http://google.com>구글</a>
  ```

<br>

### 3. **시작 태그와 종료 태그가 존재한다.**

- 대부분의 태그는 시작 태그와 종료 태그가 한 쌍으로 이루어진다.

- `<br>`태그, `<hr>`태그 등은 종료 태그가 존재하지 않는다.

  ```
  <ul>
    <li> 봄 </li>
    <li> 여름 </li>
  </ul>
  ```

<br>

### 4. **태그는 속성과 함께 사용할 수 있다.**

- HTML태그는 단순하게 태그 하나만을 사용할 수도 있으나, 속성을 이용하여 여러가지 기능을
추가할 수 있다.

- <태그 속성=“값＂속성=“값“ …….>

  `<a href=“http://www.google.com" target="_blank">구글</a>`

<br>

### 5. **적절한 들여쓰기와 내여쓰기를 사용한다.**

- HTML 태그는 기본적으로 여러칸의 공백도 한칸으로 인식

- HTML 태그 작성시 들여쓰기와 내여쓰기를 하여 가독성을 높임

- HTML 태그의 문서 구조를 확인할 때 편리함

<br>

### 6. **관계를 명확하게 표현한다.**

- HTML 태그는 태그 안에 다른 태그를 포함시킬 수 있다.

- 시작 태그와 종료 태그를 정확하게 맞춰서 사용한다.

 `<s><i> HTML는 웹페이지 작성 언어이다.</i></s>`

# HTML 태그

### 1. **제목 태그 만들기**

- `<hn>` 태그 : 일반 텍스트보다 굵고 진하게 표시 (n=1~6)

- 자동으로 위아래 여백 생성

  ```
  <hn> 제목 입력 </hn>

  <h1>제목</h1>
  본문 내용과 구별되어 표시가 되요
  ```

<br>

### 2. **주석 태그 만들기**

- HTML 문서에서 사용한 소스 코드에 대한 설명을 할 때 사용

- 주석 처리를 하면 화면에 나타나지 않는다.

  ```
  <!– 주석처리 할 내용 입력 -->

  <h1>제목</h1>
  <!– 제목 태그 중 가장 큰 글자로 표현 -->
  본문 내용과 구별되어 표시가 되요
  ```

<br>

### 3. **단락 태그**

1) `<p>` 기본구조
    
- Paragraph의 의미로 텍스트를 단락으로 만들 때 사용

- 삽입한 텍스트가 웹 브라우저 창 보다 길어질 경우 줄이 자동으로 바뀜

  ```
  <p>
  한 단락으로 구별한 텍스트
  </p>

  <h1>제목</h1>
  <p>첫번째 단락을 입력해요</p>
  <p>두번째 단락을 입력해요</p>
  ```

<br>

2) `<div>` 기본구조
    
- Division으로 입력된 텍스트의 영역을 나눌 때 사용

- 영역을 블록 단위로 묶어서 정의할 때 사용

  ```
  <div>
  블록 단위로 그룹할 내용 입력
  </div>

  <h1>제목</h1>
  <div>
  <p>첫번째 단락을 입력해요</p>
  <p>두번째 단락을 입력해요</p>
  </div>
  ```

<br>

3) `<br>` 기본구조

- 줄 바꿈을 해준다.

- 종료 태그가 존재하지 X

  ```
  텍스트 <br>

  <p>
  첫번째 단락의 내용을 입력해요<br>
  이어서 다음줄 내용도 입력해요<br>
  </p>
  ```

<br>

4) `<pre>` 기본구조

- 화면에 입력한 그대로 웹 브라우저에 표현

  ```
  <pre>
  출력한 내용 입력
  </pre>

  <pre>
  1. Html
  2. Css
  3. javascript
  </pre>
  ```

<br>

### 4. **서식 태그**

1) `<strong>`태그, `<b>`태그 기본구조

- 텍스트를 굵게 표시할 때 사용

  ```
  <strong>텍스트</strong>
  <b>텍스트</b>

  Html문서에서 작성한 텍스트 중에 <strong>굵게
  </strong>표현하고 싶은 부분에는 강조 태그를 사용하
  여 <b>굵게</b>게 표현할 수 있다.
  ```

<br>

2) `<mark>`태그 기본구조

- 형광펜 효과로 표시

  ```
  <mark>텍스트</mark>

  Html문서에 입력되어 있는 부분 중에 선택한 부분만
  배경색이 <mark>노란색</mark>으로 표시되어 하이라이
  팅 효과를 줄 수 있다.
  ```

<br>

3) `<ins>`태그, `<u>`태그 기본구조

- 텍스트 밑에 줄 형성

  ```
  <ins>텍스트</ins>
  <u>텍스트</u>

  Html문서에서 텍스트에 <ins>밑줄</ins>을 넣을 수 있
  는데 <u>밑줄</u>태그를 이용하면 텍스트 밑에 줄이
  생긴다.
  ```

<br>

4) `<sup>`태그, `<sub>`태그 기본구조

- 위 첨자와 아래 첨자를 표현

  ```
  <sup>텍</sup>스트
  <sub>텍</sub>스트

  Html문서에서도 수학식이나 화학식을 입력할 때 사용
  하는 <sup>위</sup>첨자나 <sub>아래</sub>첨자를 표
  현할 수 있다.
  ```

<br>

5) `<i>`태그, `<em>`태그 기본구조

- 기울림꼴로 표현

  ```
  <i>텍스트</i>
  <em>텍스트</em>

  Html문서에서 텍스트를 <i>비스듬<i>이 <em>기울림꼴
  </em>로 입력하면 문장에서 흐름상 특정 부분을 강조
  할 수 있다.
  ```

<br>

6) `<s>`태그, `<del>`태그 기본구조

- HTML 문서에서 삭제한 효과를 나타낸다.

  ```
  <s>텍스트</s>
  <del>텍스트</del>

  Html문서에서 실제로 <s>삭제</s>하는 것이 아니라
  <del>삭제</del>라는 의미로 취소선을 넣어서 기존의
  텍스트가 어떻게 변경되었는지 보여줄수 있다.
  ```

<br>

7) `<hr>`태그

- 수평 줄 삽입

  ```
  <hr size="굵기" width="길이
  " align="정렬방식" color="색상">

  <hr>
  Html문서에서 수평줄을 사용하면 여러 개의 텍스트 단
  락을 주제가 변할때 마다 분위기 전환용으로 사용할
  수 있다.
  <hr size=“7” width=“50%” align=“center” color =
  “blue”>
  ```

<br>

### 5. **인용문 작성하기**
  
  : 다른 사람이나 자신의 말 또는 글을 직간접적으로 끌어다 쓰는 문장으로 보통 따옴표를 사용해서 구분해 주는데, 자신의 글이 아니라는 것을 밝혀 주는 것이다.

<br>

1) `<blockquote>`태그 기본구조

- 인용문을 넣을 때 사용

  ```
  <blockquote>
  인용내용
  </blockquote>

  다른 사이트의 글을 인용할 경우 다른 텍스트와
  <blockquote>구별이 되어서</blockquote>표시가 되어
  야 한다.
  ```

<br>

2) `<cite>`태그 기본구조

- 웹 문서나 포스트에서 참고 내용 표시

  ```
  <cite>
  인용내용
  </cite>

  HTML은 웹 문서를 만들기 위하여 사용하는 기본적인
  웹 언어의 한 종류이다.
  출처 : <cite>두산백과<cite>
  ```

<br>

3) `<abbr>`태그 기본구조

- 용어를 축약형으로 표현

  ```
  <abbr title=“부연설명”>
  텍스트
  </abbr>

  <b><abbr title=“Hypertext Markup
  Language”>html</abbr></b>이란 웹 문서를 만들기 위
  하여 사용하는 기본적인 웹 언어의 한 종류이다.
  ```

<br>

## **# 경로 설정**

### @ **절대 경로**

- **http://** 로 시작하는 주소

- 인터넷상에 하나밖에 없는 유일무이한 고유한 경로를 의미

- 예 : http://image.kyobobook.co.kr/ink/images/gnb/logo_kyobo.png

<br>

### @ **상대 경로**

- 현재 폴더의 위치를 기준으로 위치 지정

- 서버의 폴더 구조나 도메인 변경 및 웹 페이지 제작에 편리

- 가장 많이 사용되는 경로

- 폴더명/이미지이름.확장자

### 6. **이미지 링크 만들기**

1) `<img>`태그 기본구조

- 웹페이지에서 지정한 페이지로 이동

  ```
  <img src = “링크 주소” 속성=“속성값”>

  <img src="http://image.kyobobook.co.kr/ink/images/gnb/logo_kyobo.png">

  <img src=“img.png” width=“300” height=“300”>
  ```

<br>

2) `<img>`태그 속성값
  
- src -> 삽입하는 이미지의 경로 지정

- width -> 이미지 너비 지정

- height -> 이미지 높이 지정

- title -> 마우스를 갖다 놓으면 이미지의 부연 설명

<br><br>

### 7. **링크 태그**

1) `<a>`태그 기본구조

- 웹페이지에서 지정한 페이지로 이동

- 속성값

  - href : 링크한 문서나 사이트의 주소 입력

  - target
    - _blank : 새 창이나 새 탭으로 열기
    - _self : 현재 페이지로 열기

  ```
  <a href = “링크 주소“, 속성=“속성값”>텍스트</a>

  <a href=“https://www.naver.com/” target=“_blank”>네이버</a>
  ```

<br>

### 8. **폼 만들기**

: 웹 페이지는 폼요소를 사용하여 사용자의 데이터를 입력받을 수 있다. <br><br>
: Input요소를 사용하면 여러 형태로 사용자의 데이터를 입력받을 수 있다. <br><br>


1) `<form>`태그 기본구조

- 폼을 만드는 기본 태그

  ```
  <form 속성=“속성값”>폼 속성 입력</form>

  <form name = “작성자 정보”>
    <input>
    <textarea>
  </form>
  ```

<br>

2) `<input>`태그 기본구조

- 입력 항목 만들기

  ```
  <input type=“유형” [속성=“속성값”]>

  <form>
    ID : <input type=“text”><br>
    비밀번호 : <input type=“password”>
  </form>
  ```

  - Type 속성 유형
    
    - text : 한줄짜리 텍스트 상자
    - password : 비밀번호 입력 필드
    - radio : 라디오 버튼
    - checkbox : 체크박스
    - file : 파일 첨부 버튼
    - submit : 서버 전송 버튼
    - botton : 버튼
    - url : url 주소 입력 필드

<br>

3) `<textarea>`태그 기본구조

- 직접 여러줄의 텍스트를 입력 가능하게 해 주는 태그

  ```
  <textarea 속성=“속성값”></textarea>

  <form>
    자기소개:
    <textarea rows=“5” cols=“80”>
    </textarea>
  </form>
  ```

<br>

### 9. **HTML5에 추가된 폼**

: HTML5에는 기존의 HTML에서 제공하던 폼 약식 외에 모바일 기기에서 활용할 수 있는 다양한 폼 양식을 제공한다. <br><br>


1) HTML5에 새롭게 추가된 폼 양식

  - type(속성)

    - color(속성값) : 색상 선택
    - datetime-local(속성값) : 날짜와 시간 선택
    - month(속성값) : 연도와 월 선택
    - number(속성값) : 숫자 선택
    - range(속성값) : 슬라이더 양식
    - Time(속성값) : 시간 선택
    - week(속성값) : 연도와 주 선택

<br>

2) `<label>`태그 기본구조

- 사용자 인터페이스의 라벨을 정의할 때 사용

  ```
  <label>텍스트</lable>

  <label>아아디</label>
  <input type=“text”>
  ```

<br>

3) `<progress>`태그 기본구조

- 작업의 진행상태를 보여줌

  ```
  <progress value=“값” [max=“값”]></progress>

  <progress value="50" max="60"></progress>
  ```

<br>

### 10. **순서 없는 목록 만들기**

1) `<ul>`태그, `<li>`태그 기본구조

- 순서가 필요하지 않은 목록 작성

  ```
  <ul>
    <li> 내용 </li>
    <li> 내용 </li>
  </ul>

  <ul type =“circle”>
    <li> 봄 </li>
    <li> 여름 </li>
  </ul>
  ```

  - 속성으로는 dics, circle, square가 있다.

### 11. **순서 있는 목록 만들기**

1) `<ol>`태그, `<li>`태그 기본구조

- 인련 번호를 부여하여 항목의 순서 작성

  ```
  <ol>
    <li> 내용 </li>
    <li> 내용 </li>
  </ol>

  <ol>
    <li> 봄 </li>
    <li> 여름 </li>
  </ol>
  ```

<br>

### 12. **정의 목록 만들기**

1) `<dl>`태그, `<dt>`태그, `<dd>`태그  기본구조

- 제목과 설명이 한 쌍인 형태로 정의 목록 작성

